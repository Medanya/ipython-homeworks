```python
import skimage
from skimage import io
import matplotlib as plt

%pylab inline
```

# 1

```python
img = io.imread('duckling.png')
io.imshow(img[:, : , :3])
io.show()
```

# 2

```python
img_upper = img[:img.shape[0]/2, :, :]
plt.imsave('ducking_upper.png', img_upper)
io.imshow(img_upper)
io.show()
```

```python
img_left = img[:, :img.shape[1]/2, :]
plt.imsave('ducking_left.png', img_left)
io.imshow(img_left)
io.show()
```

# 4

```python
mean_channels_img = img[:,:,:-1].mean(axis=2)

#    [0,1] 
color_diff = mean_channels_img.max() - mean_channels_img.min()
normal_mean_channels_img = (mean_channels_img - mean_channels_img.min()) / color_diff

gray_img = skimage.color.rgb2gray(img)
concat_img = np.concatenate((normal_mean_channels_img, gray_img), axis=1)

print("    | rgb2gray")
io.imshow(concat_img)
plt.show()
```
