```python
import hashlib
```

```python
sha_set = set()

for i in xrange(50000000):
    value = hashlib.sha256(str(i)).digest()[:6]
    if value in sha_set:
        print i, value
        break
    else:
        sha_set.add(value)
```

```python
target = hashlib.sha256(str(47618354)).digest()[:6]
target
```

```python
for i in xrange(50000000):
    if hashlib.sha256(str(i)).digest()[:6] == target:
        print i
        break
```
